Решать задачу будем методом линейной регрессии. 

Общая идея: имеем два фьючерса $f_1(t), f_2(t)$ BTC и ETH соответственно. Нахождение собственного движения второго цены фьючерса 2 представимо в виде

$$
f_2(t) = \varphi(f_1(t)) + y_2(t)
$$
где y_2(t) – интересующее нас собственное движениею.
Его и находим по методу линейной регрессии, отысканием коэффициентов.

```python
last_time =  datetime.datetime.now()  
last_eigen_price = 635.9 

time_ei = [] 
price_ei = []
while True:
 
  now = datetime.datetime.now()  

  print(now)  
  start = now - datetime.timedelta(seconds = 1000*60*60) 
  end = now + datetime.timedelta(seconds = 10)  

  data1 = yf.download(['BTC-USD'], start = start, end = end, interval="2m",show_errors=True) 
  data2 = yf.download(['ETH-USD'], start = start, end = end, interval="2m", show_errors=False)
   
  BTC_USD_time = data1["Adj Close"].dropna().index.to_list() #.pct_change()
  BTC_USD_price = data1["Adj Close"].dropna().to_list() #USDT-ETH  

  ETH_USD_time  = data2["Adj Close"].dropna().index.to_list() 
  ETH_USD_price = data2["Adj Close"].dropna().to_list()
       
  print("correl until preprocesing",np.corrcoef(BTC_USD_price,ETH_USD_price)[0][1])  

  X = pd.DataFrame({'x': BTC_USD_price, 'c': [1]*len( BTC_USD_price)}, index = range(len(BTC_USD_price))) 
  Y = ETH_USD_price  

  y = Y
  reg = LinearRegression().fit(X.to_numpy(), y)  
  eigen_price = np.array(ETH_USD_price)  -  reg.coef_[0]*np.array(BTC_USD_price )
  print("correl eigen", np.corrcoef(np.array(BTC_USD_price),   eigen_price )[0][1]) 
   
   
  if abs(eigen_price[-1] -  last_eigen_price)/last_eigen_price > 0.01:   
    print(eigen_price[-1] )
    print("/n")
    print("delta PRICE > 1%") 
    print("/n")

  if now - last_time > datetime.timedelta(seconds = 60*60): 
    last_time = now  
    last_eigen_price = eigen_price[-1]  

  if len(time_ei ) ==0:
    time_ei += [now ] 
    price_ei += [eigen_price[-1]]  
  else: 
    if  eigen_price[-1] != price_ei[-1]:  
      time_ei += [now ] 
      price_ei += [eigen_price[-1]]  

      fig = go.Figure()
      #fig.add_trace(go.Scatter(x=BTC_USD_time, y=data1["Adj Close"].dropna().to_list()))
      fig.add_trace(go.Scatter(x=time_ei ,y=price_ei )) 
      fig.add_trace(go.Scatter(x=[last_time] ,y=[last_eigen_price ] ))
      fig.show() 

   


  time.sleep(10)
```
